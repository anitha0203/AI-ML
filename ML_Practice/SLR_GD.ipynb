```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataFiles/advertising.csv")
print(data)
```

```python
x = data.iloc[:, 0]
y = data.iloc[:, 1]
plt.scatter(x, y)
plt.show()
```

```python
m = 0
c = 0

l = 0.00001
iterations = 1000


for i in range(iterations):
  y_pred = m * x + c
  grad_m = (1/len(x)) * sum((y_pred - y) * x)
  grad_c = (1/len(x)) * sum(y_pred - y)

  m = m - l * grad_m
  c = c - l * grad_c

print(m)
print(c)

y_pred = m * x + c
print(y_pred)
```

```python

plt.scatter(x, y) 
plt.plot(x, y_pred, color='red')  # regression line
plt.show()
```
