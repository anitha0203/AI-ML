```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataFiles/employees.csv")
```

```python
x = data['YearsExperience'].values
y = data[' Salary'].values
```

```python
x_mean = np.mean(x)
y_mean = np.mean(y)
```

```python
# Finding the m, c and y_pred values
num = 0
den = 0

for i in range(len(x)):
    num += (x[i] - x_mean ) * (y[i] - y_mean)
    den += (x[i] - x_mean) ** 2

m = num / den
c = y_mean - m * x_mean
y_pred = m * x + c
```

```python
plt.scatter(x, y, color='red')
plt.plot(x, y_pred, color='blue')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
ss_t = 0
ss_r = 0

for i in range(len(x)): 
  y_pred = c + m * x[i]
  ss_t += (y[i] - y_mean) ** 2
  ss_r += (y[i] - y_pred) ** 2
r2 = 1 - (ss_r/ss_t)

print(r2)
```
