```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X = 2 * np.random.rand(50, 1) 
y = 4 + 3 * X + np.random.randn(50, 1) 
print(X)
print(y)
```

```python
plt.scatter(X, y)
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression Example")
plt.show()
```

```python
m = 0 
c = 0 
learning_rate = 0.01
iterations = 1000


for i in range(iterations):
    y_pred = m * X + c 
    grad_m = (-2 / len(X)) * sum(X * (y - y_pred))
    grad_c = (-2 / len(X)) * sum(y - y_pred)
    m -= learning_rate * grad_m
    c -= learning_rate * grad_c

print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
y_pred = m * X + c

plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.xlabel("X")
plt.ylabel("y")
plt.title("Simple Linear Regression with Gradient Descent")
plt.show()
```
