```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data with three features
np.random.seed(0)
X1 = 2 * np.random.rand(100, 1)
X2 = 3 * np.random.rand(100, 1)
X3 = 4 * np.random.rand(100, 1)
y = 4 + 3 * X1 + 2 * X2 + 1.5 * X3 + np.random.rand(100, 1)
```

```python
# Concatenate the features
X = np.c_[X1, X2, X3]

# Add a column of ones for the bias term
X_b = np.c_[np.ones((100, 1)), X]

# Initialize random coefficients for our model
theta = np.random.randn(4, 1)
```

```python
# Initialize random coefficients for our model
theta = np.random.randn(4, 1)

# Set hyperparameters
learning_rate = 0.01
n_iterations = 1000

# Perform gradient descent
for iteration in range(n_iterations):
    gradients = -2 / len(X) * X_b.T.dot(y - X_b.dot(theta))
    theta -= learning_rate * gradients

# Print the learned coefficients
print("Learned coefficients:", theta)
```

```python
# Create a scatter plot of the data points
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X1, X2, y)

# Generate predictions based on the learned model
y_pred = X_b.dot(theta)

# Plot the regression plane
xx1, xx2 = np.meshgrid(X1, X2)
yy_pred = theta[0] + theta[1] * xx1 + theta[2] * xx2
ax.plot_surface(xx1, xx2, yy_pred, alpha=0.5)

ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')
plt.show()
```
