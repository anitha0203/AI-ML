```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv("../dataFiles/salaries.csv")

X = data.iloc[:, 1:-1].values
y = data.iloc[:, -1].values

# Add polynomial features (e.g., degree=4)
degree = 4
X_poly = np.column_stack([X ** i for i in range(1, degree + 1)])

# Normalize features (optional but recommended)
X_poly = (X_poly - X_poly.mean(axis=0)) / X_poly.std(axis=0)
y = (y - y.mean()) / y.std()

# Initialize coefficients
m = np.zeros(degree)
learning_rate = 0.1
epochs = 10000

for epoch in range(epochs):
    y_pred = X_poly.dot(m)
    error = y_pred - y
    gradients = (2 / len(X)) * X_poly.T.dot(error)
    m -= learning_rate * gradients

# Visualize the polynomial regression
plt.scatter(X, y, color="red")
X_fit = np.linspace(X.min(), X.max(), 100)
X_fit_poly = np.column_stack([X_fit ** i for i in range(1, degree + 1)])
y_fit = X_fit_poly.dot(m)
plt.plot(X_fit, y_fit, color="blue")
plt.xlabel('Position Level')
plt.ylabel('Salary')
plt.show()

# Make a prediction for a specific input
input_value = 2.5
input_poly = np.array([input_value ** i for i in range(1, degree + 1)])
input_poly = (input_poly - X_poly.mean(axis=0)) / X_poly.std(axis=0)
predicted_salary = input_poly.dot(m)
predicted_salary = (predicted_salary * y.std()) + y.mean()

print(f"Predicted Salary for Position Level {input_value}: {predicted_salary}")
```
