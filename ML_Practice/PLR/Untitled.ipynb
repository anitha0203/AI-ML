```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 3)  # 3 features
y = 2 + X[:, 0] + 0.5 * X[:, 1] + 2 * X[:, 2] + np.random.randn(100)
```

```python
# Polynomial regression using scikit-learn
degree = 2  # degree of the polynomial
model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
model.fit(X, y)

# Print the coefficients
print("Intercept:", model.named_steps['linearregression'].intercept_)
print("Coefficients:", model.named_steps['linearregression'].coef_)
```

```python
# Make predictions
y_pred = model.predict(X)

# Evaluate the model
mse = mean_squared_error(y, y_pred)
print("Mean Squared Error:", mse)
```

```python
# Plot the results
plt.scatter(X[:, 0], y, label='Actual')
plt.scatter(X[:, 0], y_pred, label='Predicted')
plt.xlabel('Feature 1')
plt.ylabel('Target')
plt.legend()
plt.show()
```
