```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("dataFiles/advertising.csv")
print(data)
```

```python
X = data.iloc[:, 0]
y = data.iloc[:, 1]
plt.scatter(X, y)
plt.show()
```

```python
m = 0 
c = 0 
learning_rate = 0.001
epochs = 1000

for i in range(epochs):
    y_pred = m * X + c 

    grad_m = (-2 / len(X)) * sum(X * (y - y_pred))
    grad_c = (-2 / len(X)) * sum(y - y_pred)
    print(grad_m)

    m -= learning_rate * grad_m
    c -= learning_rate * grad_c

print("Slope (m):", m)
print("Intercept (c):", c)
```

```python

plt.scatter(X, y)
plt.plot(X, m * X + c, color='red')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```
